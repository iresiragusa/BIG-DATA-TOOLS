Pandas is a comprehensive library for processing and analyzing multidimensional data. It is based on two main data structures which are the `Series` and the `DataFrame`.

A `Series` represents a 1D array of data homogeneous in type, while a `DataFrame` is a two-dimensional organization of data organized by columns of heterogeneous type.

Columns can be removed and inserted, and, in general, a dataframe allows for simple groupby, merge, slicing, data indexing, reshaping operations, and allows for handling loading and saving in `csv`, `json`, `SQL`, `HTML`, `Excel` and others.

Mathematical and statistical operations are available on dataframes and apply to all cells or to a user-specified subset.

```python
# https://pandas.pydata.org/docs/reference/index.html

import numpy as np
import pandas as pd
import random

# examples of series

a = pd.Series([7, 11, 13, 17])
print(a)
b =  pd.Series(100.0, range(5))
#print(b)
c = pd.Series([random.randint(0, 100) for x in range(20)])
#print(c)
d = pd.Series([98.6, 98.9, 100.2, 97.9], index=['Julie', 'Charlie', 'Sam', 'Andrea'])
#print(d)
```

```python
# create a simple dataframe, from a dict of list,
# each of them will be a series

data = {
    'apples': [3, 2, 0, 1],
    'oranges': [0, 3, 7, 2]
}

purchases = pd.DataFrame(data)

purchases
```

```python
# we can add a list of indexs to our data

purchases = pd.DataFrame(data, index=['June', 'Robert', 'Lily', 'David'])

purchases
```

```python
# find a value
# loc w/ index name
# iloc w/ index value
print(purchases.loc[['June', 'Lily']])
print(purchases.iloc[[0, 2]])

print('------')

print(purchases.loc['June','oranges'])
print(purchases.iloc[0, 1])
```

```python
# find a value matrix-like
# at w/ index name
# iat w/ index value
print(purchases.at['June','oranges'])
print(purchases.iat[0, 1])
```

```python
# drop cabin attribute
titanic.drop('Cabin', axis=1, inplace=True)

titanic
```

```python
titanic.groupby(['Sex', 'Pclass'])['Age'].agg(['mean', 'count'])
```

```python
mean_age_female_3pclass = titanic.groupby(['Sex', 'Pclass'])['Age'].mean().at['female', 3]

print(mean_age_female_3pclass.round())
```

```python
def fill_age_with_mean(titanic):
    aggregated = titanic.groupby(['Sex', 'Pclass'])['Age'].mean()
    for i, row in titanic.iterrows():
        if pd.isnull(row['Age']):
            mean_age = aggregated.at[row.Sex, row.Pclass].round()
            titanic.at[i, 'Age'] = mean_age

fill_age_with_mean(titanic)
```

```python
age_labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
titanic['age_group'] = pd.cut(titanic.Age, range(0, 81, 10), right=False, labels=age_labels)
titanic
```

```python
print(titanic.groupby(['age_group']).size())
titanic.groupby(['age_group']).size().plot.bar()
```
