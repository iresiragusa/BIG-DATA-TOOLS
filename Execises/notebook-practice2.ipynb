`List Comprehension` the python-way to create dynamically a list
 
`newlist = [expression for item in iterable if condition == True]`

```python
# classic way
fruits = ["apple", "banana", "cherry", "kiwi", "lemom", "mango"]
newlist = []

for x in fruits:
  newlist.append(x)

print(newlist) 

# list comprehension
newlist_2 = [x for x in fruits]
print(newlist_2)
```

```python
# things can be a bit more complex with some conditions

# classic way
newlist = []

#enumerate outputs the index and the value
for idx, x in enumerate(fruits):
    if idx%2==0:
        newlist.append(x)

print(newlist) 

# list comprehension
newlist_2 = [x for idx, x in enumerate(fruits) if idx%2==0]
print(newlist_2)
```

`Dictionary Comprehension` the python-way to create dynamically a dictionary

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

dict2 = {}

for k,v in dict1.items():
    dict2[k] = v**2

print(dict2)

# dict comprehension
dict3 = {k:v**2 for k,v in dict1.items()}
print(dict3)
```

`NumPy` is the algebric computing library that implements vectors and arrays of numeric type and operations on them. The n-dimensional array (`ndarray`) is structured as an object of a certain type and its own layout in memory. The `ndarray` is characterized by its `shape` that is the tuple of its dimensions.

```python
# https://numpy.org/doc/stable/reference/index.html#reference

import numpy as np

a = np.array([[1,2,3],[4,5,6]])
# optionally specify dtype='int64' or 'float64'

print(a)
print(a[0][0], type(a[0][0]))
print('shape ->', a.shape)

b = np.array(a,dtype='float64')

print(b)
print(b[0][0], type(b[0][0]))

print('shape ->', b.shape)
```

```python
# arrays of zeros
z = np.zeros((2,3))
print('zeros')
print(z)

e = np.eye(4,dtype='int32')
print('diag')
print(e)

fl = np.full_like(a,3.6)
print('full like')
print(fl)

f = np.full((3,2),4)
print('full')
print(f)

```

```python
print(a)

print('value', a[1,2])

print('\n')

# arrays uses the native Python slicing: working on a slice means change the original array
print('slicing\n', a[:,1:])
print('\n')
print('a shape\n', a.shape)
print('\n')
# array w/ even values
print('even values\n', a[a%2==0])
print('\n')

print('transpose property\n', a.T) # as property (no modification)
print('transpose method\n', a.transpose()) # as method (no modification)
print('\n')
print('reshape\n', a.reshape(3,2)) # reshaping: restituisce un array con gli elemen
```

```python
# member-member operators
print('a+b\n', a+b, '\n')

print('a*b\n', a*b, '\n')

print('a<=3\n', a<=3, '\n')
```

```python
# range generator
# np.arange(start,stop,step,dtype=None)
# stop is excluded
x = np.arange(-1,1.1,0.1)

print(type(x[0]))
print(x)

# print formattation
np.set_printoptions(precision=3, suppress=True)

print(x)
```

```python
# linear algebra operator

x=np.array([1,2,3])
y=np.array([4,5,6])
m = np.array([[1,2,3],[7,10,9],[4,5,6]])
n = np.array([[2,4],[6,8],[10,12]])

print('x\n', x)
print('y\n', y)
print('m\n', m)
print('n\n', n)

# matrix product
print('# matrix product')
print(np.dot(x,y),end='\n\n')
print(np.dot(m,n),end='\n\n')

# outer product
print('# outer product')
print(np.outer(x,y),end='\n\n')
print(np.outer(m,n),end='\n\n')

# determinant
print('# determinant')
print(np.linalg.det(m),end='\n\n')
```

`Matplotlib` is a library in which a API called `pyplot` can be useful for plot graphs and add legend and graphics

```python
# https://matplotlib.org/stable/api/index.html

import math as m
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.ylim(0,6)

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
# subplots

data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(11, 7), sharey=True)
axs[0].bar(names, values, color='red')
axs[1].scatter(names, values, marker='x', color='magenta')
axs[2].plot(names, values, color='cyan')

fig.suptitle('Mixed Plotting')

# show graphs
plt.show()
```
